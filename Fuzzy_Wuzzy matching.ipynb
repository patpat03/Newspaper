```python
from fuzzywuzzy import process, fuzz
import pandas as pd
```

```python
xml = "/Users/Mihir/Downloads/Sample/dronepatents_assignees.csv"
df = pd.read_csv(xml)
```

```python
org = df['organization'].tolist()
```

```python
temp = []
#Removes duplicates
org = set(org)
for i in org:
    temp.append(i)
```

```python
df = {}
#Uses partial ratio to find most similar substrings for each of the companies in the csv 
for i in org:
    temp.remove(i)
    r = process.extract(i,temp)
    df[i] = r
    temp.append(i)
```

```python
print(df)
```

```python
one = []
two = []
three = []
#creates array in a format that can be easily read by the csv file
for i in df.keys():
    x = []
    x.append(i)
    x.append(df[i])
    one.append(x)
```

```python
dataframe=pd.DataFrame(one, columns=["company", 'match']) 
```
